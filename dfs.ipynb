```python
import matplotlib.pyplot as plt
import networkx as nx

Graph_nodes = {
    'A': ['B', 'C', 'D'],
    'B': ['E', 'F'],
    'C': [],
    'D': ['G', 'H', 'I'],
    'E': ['J'],
    'F': ['K'],
    'G': ['L'],
    'H': [],
    'I': [],
    'J': [],
    'K': [],
    'L': []
}

visited = set()

def dfs(visited, Graph_nodes, node):
    if node not in visited:
        print(node, end=" ")
        visited.add(node)
        for neighbor in Graph_nodes[node]:
            dfs(visited, Graph_nodes, neighbor)

print("The D.F.S. is: ")
dfs(visited, Graph_nodes, 'A')

def draw_graph(visited):
    G  = nx.Graph()
    for node, edges in Graph_nodes.items():
        for neighbor in edges:
            G.add_edge(node, neighbor)
    pos = nx.spring_layout(G)
    nx.draw(G, pos, with_labels = True, node_color = 'lightblue', node_size = 2000, font_size = 16)
    visited_color = ['lightgreen' if node in visited else 'lightblue']
    nx.draw_networkx_nodes(G, pos, node_color = visited_color, node_size = 2000)
    plt.title("Graph Visualization with D.F.S Traversal Highlighted")
    plt.show()
draw_graph(visited)
```
