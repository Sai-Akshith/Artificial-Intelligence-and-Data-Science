```python
import matplotlib.pyplot as plt
import networkx as nx

Graph_nodes = {
    'A': ['B', 'C', 'D'],
    'B': ['E', 'F'],
    'C': [],
    'D': ['G', 'H', 'I'],
    'E': ['J'],
    'F': ['K'],
    'G': ['L'],
    'H': [],
    'I': [],
    'J': [],
    'K': [],
    'L': []
}
visited = []
queue = []
def bfs(visited, Graph_nodes, node):     # node has the initial node in it i.e., A for the first time
    visited.append(node)
    queue.append(node)
    while queue:                         # until the queue is empty continue in the loop
        m = queue.pop(0)
        print(m, end = " ")
        for neighbor in Graph_nodes[m]:
            if neighbor not in visited:
                visited.append(neighbor)
                queue.append(neighbor)
print("The B.F.S. is: ")
bfs(visited, Graph_nodes, 'A')

def draw_graph(visited):
    G = nx.Graph()
    for node, edges in Graph_nodes.items():
        for neighbor in edges:
            G.add_edge(node, neighbor)
    pos = nx.spring_layout(G)
    nx.draw(G, pos, with_labels = True, node_color = 'lightblue', node_size = 2000, font_size = 16)
    visited_color = ['lightgreen' if node in visited else 'lightblue' for node in G.nodes()]
    nx.draw_networkx_nodes(G, pos, node_color = visited_color, node_size = 2000)
    plt.title("Graph Visualization with BFS Traversal Highlighted")
    plt.show()
draw_graph(visited)
```
